3.2 Continuity and Differentiation 
Continuity in a function is defined as no abrupt changes in value. 
We can compare the example below in continuous function y=2x vs a non-continuous function y=1/x

```python
import numpy as np
x=np.linspace(-4,4,50)
yone=2*x
ytwo=1/x
import matplotlib.pyplot as plt 
plt.figure
plt.plot(x, yone,'-r') 
plt.plot(x, ytwo,'-b') 
plt.title('Continuity') 
```

Rolle theorm:  Let f : [a,b] → R be a continuous function and assume that its derivative exists on (a,b). If f(a)=f(b), then there is a<c< b such
that f′(c) = 0.

Mean value theorm: Let f:[a,b]→R be a continuous function and assume that its derivative exists on (a,b). Then there is a<c< b such that f(b)=f(a)+(b−a)f′(c)

Partial derivative is when you take a derivative with respect to one variable while holding the other variables constant. We will explore an example of partial derivatives below.

```python
from sympy import symbols, diff
x, y, z = symbols('x y z', real=True)
f = 12*x*y*z + x*cos(z) + x**2 + z**8*y
partial_x=diff(f, x)
partial_y=diff(f,y)
partial_z=diff(f,z)
print("partial_x=",partial_x)
print("partial_y=",partial_y)
print("partial_z=",partial_z)
```

Taylor's theorm
-Taylor’s theorem gives an approximation of a differentiable function around
a given point by a polynomial
-  Let f : D → R where D ⊆ R. Suppose f has a m times continuous derivative of[a,b]. Then f(b) = f(a) + (b−a)f′(a) +0.5(b−a)^2 f′′(a)+...+((b−a)^(m−1))/(m−1)! f^(m−1)(a)+Rm. Where Rm = (1/m!)f^m (a+θ(b-a)) for 0 < θ < 1.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

x = np.linspace(-np.pi, np.pi, 200)
y = np.zeros(len(x))

labels = ['First Order', 'Third Order']

plt.figure(figsize = (10,8))
for n, label in zip(range(2), labels):
    y = y + ((-1)**n * (x)**(2*n+1)) / np.math.factorial(2*n+1)
    plt.plot(x,y, label = label)

plt.plot(x, np.sin(x), 'k', label = 'Analytic')
plt.grid()
plt.title('Taylor Series Approximations of Various Orders')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
